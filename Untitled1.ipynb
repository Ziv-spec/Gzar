```python

# add keywords as needed into here
keywords = [
        ["bool",   "TK_BOOL_TYPE"],
        ["cast",   "TK_CAST"],
        ["else",   "TK_ELSE"],
        ["false",  "TK_FALSE"],
        ["if",     "TK_IF"],
        ["int",    "TK_INT_TYPE"],
        ["nil",    "TK_NIL"],
        ["return", "TK_RETURN"],
        ["s16",    "TK_S16_TYPE"],
        ["s32",    "TK_S32_TYPE"],
        ["s64",    "TK_S64_TYPE"],
        ["s8",     "TK_S8_TYPE"],
        ["string", "TK_STRING_TYPE"],
        ["true",   "TK_TRUE"],
        ["u16",    "TK_U16_TYPE"],
        ["u32",    "TK_U32_TYPE"],
        ["u64",    "TK_U64_TYPE"],
        ["u8",     "TK_U8_TYPE"],
        ["void",   "TK_VOID_TYPE"],
        ["while",  "TK_WHILE"],
]

buff = list(range(1, 26)) # character buffer for a-z characters
used_characters = [0]*26  # buffer for characters which were used before
mapbuff = [0]*32          # final buffer which will be the actual hashmap

def calc_hash(string):
    value_hash = 0
    for c in string:
        index = abs(ord(c)-ord('a'))%26
        used_characters[index] += 1
        value_hash += buff[index]
    return value_hash

collided_strings = []

for string, tk in keywords:
    value_hash = calc_hash(string)
    mapi = value_hash%len(mapbuff)

    if mapbuff[mapi] == 1:
        collided_strings.append(string)
    else:
        mapbuff[mapi] = 1
        
```

```python
for k in range(10):
    j = 0
    for i, istaken in enumerate(mapbuff):
        if not istaken: 
            print("collided strings:", collided_strings)
            if not collided_strings: 
                break;
            # compute hash for a string
            string = collided_strings[j]
            value_hash = calc_hash(string)
            mapi = value_hash%len(mapbuff)
            
            # check which character was used the least amount of times 
            # up til this point
            min_used_character = 10000000000
            min_used_character_index = -1
            for c in string: 
                index = abs(ord(c)-ord('a'))%26
                if used_characters[index] < min_used_character:
                    min_used_character = used_characters[index]
                    min_used_character_index = index
            j += 1
            
            if min_used_character_index > -1:
                buff[min_used_character_index] += mapi-i

                # recompute hashes
                collided_strings = []
                mapbuff = [0]*32
                for string, tk in keywords:
                    value_hash = calc_hash(string)
                    mapi = value_hash%len(mapbuff)

                    if mapbuff[mapi] == 1:
                        collided_strings.append(string)
                    else:
                        mapbuff[mapi] = 1
                j=0
            
```

```python
# validate that I don't have collisions with the values of the strings and thier hashes

hashes = [[0, 0, 0]]*32
for string, tk in keywords:
    value_hash = calc_hash(string)
    mapi = value_hash%len(mapbuff)
    if value_hash in hashes:
        print('bad')
    hashes[mapi] = [value_hash, string, tk]
    print(value_hash, mapi)
#hashes
```
